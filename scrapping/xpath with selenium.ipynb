```python
# Se utiliza selenium para dar click en el boton debido a que es una pagina dinamica
import time
from lxml import html
import pandas as pd
from selenium import webdriver

#Se levanda un webdriver
options = webdriver.ChromeOptions()
options.add_argument("start-maximized"); # open Browser in maximized mode
options.add_argument('--headless')
options.add_argument("--disable-infobars"); # disabling infobars
options.add_argument("--disable-extensions"); # disabling extensions
options.add_argument("--disable-gpu"); # applicable to windows os only
options.add_argument("--disable-dev-shm-usage"); # overcome limited resource problems
options.add_argument("--no-sandbox"); # Bypass OS security model
options.add_argument('--remote-debugging-port=9222')
chrome_driver = "/tmp/chromedriver/chromedriver"
driver = webdriver.Chrome(executable_path=chrome_driver, chrome_options=options)

#Se ejecutan los pasos correspondientes para llegar a la tabla
root_link = "https://en.wikipedia.org"
first_link = "/wiki/Kevin_Bacon"
driver.get(root_link + first_link)
```

```python
page = html.fromstring(driver.page_source)

elements = page.xpath("//a[@href]")
links = []
for element in elements:
    href = element.get("href")
    if(href.find("/wiki/") != -1):
        links.append(element.get("href"))

LAST_MODIFIED = page.xpath('//*[@id="footer-info-lastmod"]/text()')
TITLE = page.xpath('//*[@id="firstHeading"]/text()')
print( root_link + first_link, TITLE, LAST_MODIFIED)

for link in links:
    driver.get(root_link + link)
    page = html.fromstring(driver.page_source)
    
    LAST_MODIFIED = page.xpath('//*[@id="footer-info-lastmod"]/text()')
    TITLE = page.xpath('//*[@id="firstHeading"]/text()')
    
    print( root_link + link, TITLE, LAST_MODIFIED)
```
